```python
file = open("C:/Users/U0097901/Census10000arct.txt", "r")
content = file.read()
file.close()
x=content.split("\n")

L=[]
M=""
for t in x[17:]:
    if len(t)>0:
        M+=t
        if t[-1]==";":
            L.append(M)
            M=""
import re
Pol=[]
for M in L:
    for i in range(6,300):
        if M[i:i+7]==" | {   ":
            #Pol.append(M[i+4:-4]);
            X=re.split('}, {|},{|},  {|,', M[i+8:-6])
            Pol.append(X)
            break
G=[]
for X in Pol:
    g=Graph();
    for j in range(0,len(X),2):
        a=int(X[j]);
        b=int(X[j+1]);
        g.add_edge((a,b));
    G.append(g)
```
