```python
L=[]
M=""
for t in x[17:]:
    if len(t)>0:
        M+=t
        if t[-1]==";":
            L.append(M)
            M=""
```

```python
import re
Pol=[]
for M in L:
    for i in range(6,300):
        if M[i:i+7]==" | {   ":
            #Pol.append(M[i+4:-4]);
            X=re.split('}, {|},{|},  {|,', M[i+8:-6])
            Pol.append(X)
            break
```

```python
G=[]
for X in Pol:
    g=Graph();
    for j in range(0,len(X),2):
        a=int(X[j]);
        b=int(X[j+1]);
        g.add_edge((a,b));
    G.append(g)
```

```python
import time
x=time.time()
Lv=[]
for g1 in G[-2:-1]:
    g=g1.copy()
    g.subdivide_edges(g.edges(labels=False),1);
    g=g.line_graph();
    W=g.wiener_index();
    g.delete_vertex(g.vertices()[0]);
    W2=g.wiener_index()
    if g1.order()>1000:
        Lv.append(W-W2)
    if abs(W-W2)<35:
        print(g1.order())
y=time.time()
print(y-x)
```
